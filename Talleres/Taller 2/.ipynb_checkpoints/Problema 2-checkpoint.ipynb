```python
from deap import algorithms, base, creator, tools
import random
import numpy as np
```

```python
creator.create("FitnessMin", base.Fitness, weights=(-1.0,)) # -1 -> minimum problem
creator.create("Individual", list, fitness=creator.FitnessMin)
```

```python
toolbox = base.Toolbox()
```

```python
def obtIndividuo(): # Funcion para generar un atributo del individuo can valores entre -5 y 5
    x= random.uniform(-5,5)
    return x
```

```python
indv = obtIndividuo()
indv
```

```python
# registro del generador de atributos
toolbox.register("attr_bool", obtIndividuo)

```

```python
#prueba de generador de atributos
c=toolbox.attr_bool()
print(c)
```

```python
# Inicializador de estructuras
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, 2)
#Se establece que el induviduo se genera con 2 atributos
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

```python
#prueba de generador de poblacion
d=toolbox.population(n=5)
print(d)
```

```python
#Funci√≥n de evaluaci√≥n
#Retorna el valor de z, segun la ecuacion dada en el ejercicio

def eval(individual):
    i1= individual[0] 
    i2= individual[1] 
    x1=2.1*(i1**2)
    x2=4-x1
    x3=x2+((i1**4)/3)
    x4=x3*(i1**2)
    x5=i1*i2
    x6=(-4+(4*(i2**2)))
    x7=x6*(i2**2)
    z=x4+x5+x7
    return z,
```

```python
# registro de la funcion de evaluacion
toolbox.register("evaluate", eval)

```

```python
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutUniformInt, low=-5.00, up=5.00, indpb=0.3)
toolbox.register("select", tools.selTournament, tournsize=3)
```

```python
#creacion de poblacion
pop = toolbox.population(n=300)
#pop
```

```python
#Uso de un HallOfFame para hacer un seguimiento del mejor individuo que aparece en la evoluci√≥n
hof = tools.HallOfFame(1)
```

```python
# Estadisticas
stats = tools.Statistics(lambda ind: ind.fitness.values)
stats.register("avg", np.mean)
stats.register("std", np.std)
stats.register("min", np.min)
stats.register("max", np.max)
```

```python
# algorithms : contiene implementaciones utiles de algunos AG basicos
pop, log = algorithms.eaSimple(pop, toolbox, cxpb=0.5, mutpb=0.3, ngen=200, stats=stats, halloffame=hof, verbose=True)
```

```python
b=tools.selBest(pop, k=5)
#print(b)
```

```python
for elem in b:
    print(eval(elem))
```
